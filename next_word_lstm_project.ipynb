```python
import re


def remove_emojis(text):
    # Define the emoji pattern to include sequences as well
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Extended-A
                               u"\U0001FA70-\U0001FAFF"  # Extended-B
                               u"\u2600-\u26FF"          # Miscellaneous Symbols
                               u"\u2700-\u27BF"          # Dingbats
                               u"\u2B50"                 # Star
                               u"\u2934-\u2935"          # Arrows
                               u"\u2B05-\u2B07"
                               u"\u2B1B"                 # Black Large Square
                               u"\u3030"                 # Wavy Dash
                               u"\u303D"                 # Part Alternation Mark
                               u"\u3297"                 # Circled Ideograph Congratulation
                               u"\u3299"                 # Circled Ideograph Secret
                               u"0-9\ufe0f\u20e3"       # Digits and Combining Enclosing Keycap
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)


# Split the data into lines
lines = github_copilot_faqs.split('\n')

# Remove emojis from each line
cleaned_lines = [remove_emojis(line) for line in lines]

# Join the cleaned lines back into a string
cleaned_data = '\n'.join(cleaned_lines)

# Print the cleaned data
print(cleaned_data)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([cleaned_data])
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []
for sentence in cleaned_data.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
len(tokenizer.word_index)
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=376)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(376, 100, input_length=max_len - 1))
model.add(LSTM(200))
model.add(Dense(376, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)
```

```python
import time
import numpy as np

text = "what is Github copilot"

for i in range(20):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=max_len-1, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```

```python
import ipywidgets as widgets
from IPython.display import display

# Define the function to preprocess and predict the next word
def predict_next_word(change):
    # Get the text from the input field
    text = change.new

    # Remove emojis
    text = remove_emojis(text)

    # Tokenize
    token_text = tokenizer.texts_to_sequences([text])[0]
    # Padding
    padded_token_text = pad_sequences([token_text], maxlen=max_len - 1, padding='pre')
    # Predict
    pos = np.argmax(model.predict(padded_token_text))

    for word, index in tokenizer.word_index.items():
        if index == pos:
            # Append the predicted next word to the input text
            output_text.value = text + " " + word

# Create widgets
text_input = widgets.Text(description="Enter a sentence:")
output_text = widgets.Text(description="Next Word:", disabled=True)

# Attach event handler to observe changes in the text input
text_input.observe(predict_next_word, names='value')

# Display widgets
display(text_input)
display(output_text)

```
